```python
import pandas as pd
import numpy as np

# Load the dataset from the provided file path
file_path = 'shipment_sample_data_1.csv'
df = pd.read_csv(file_path)

# Filter the data for 'CNNGB' in 'port_of_arr'
df = df[df['port_of_arrival'] == 'CNNGB']

# Convert arrival and departure times to datetime
df['arrival_time'] = pd.to_datetime(df['arrival_time'], format='%d-%m-%Y %H:%M')
df['depart_time'] = pd.to_datetime(df['depart_time'], format='%d-%m-%Y %H:%M')

# Calculate transit time in hours
df['transit_time'] = (df['arrival_time'] - df['depart_time']).dt.total_seconds() /( 3600*24)

# Display the first few rows of the dataset
df
```

```python
# Define the bootstrapping function for the entire dataset
def bootstrap_sample_df(dataframe, n_samples):
    bootstrapped_samples = []
    for _ in range(n_samples):
        sample = dataframe.sample(frac=1, replace=True)
        bootstrapped_samples.append(sample)
    return bootstrapped_samples
    
```

```python
# Number of bootstrap samples
n_samples = 10

# Perform bootstrapping on the entire dataset
bootstrapped_dataframes = bootstrap_sample_df(df, n_samples)

# Combine bootstrapped samples into one DataFrame for easier manipulation
bootstrapped_df_combined = pd.concat(bootstrapped_dataframes)

# Display the first few rows of the combined bootstrapped DataFrame
bootstrapped_df_combined
```

```python
# Store the bootstrapped dataset in a variable
bootstrapped_df = bootstrapped_df_combined

# Display the first few rows of the bootstrapped dataset variable to confirm
bootstrapped_df

```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = bootstrapped_df['transit_time'].quantile(0.25)
Q3 = bootstrapped_df['transit_time'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define the lower and upper whiskers
lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR

# Remove outliers
df_cleaned = bootstrapped_df[(bootstrapped_df['transit_time'] >= lower_whisker) & (bootstrapped_df['transit_time'] <= upper_whisker)]

# Display the first few rows of the cleaned dataset
df_cleaned
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Assuming df is your cleaned dataset with 'scac' and 'transit_time'

# Aggregate data by carrier to get shipment_count and mean_transit_time
carrier_stats = df.groupby('scac').agg(
    shipment_count=('transit_time', 'size'),
    mean_transit_time=('transit_time', 'mean')
).reset_index()

# Prepare the data for regression
X = carrier_stats['shipment_count'].values.reshape(-1, 1)  # Independent variable
y = carrier_stats['mean_transit_time'].values  # Dependent variable

# Fit the linear regression model
linear_regressor = LinearRegression()
linear_regressor.fit(X, y)

# Make predictions
y_pred = linear_regressor.predict(X)

# Calculate coefficients
intercept = linear_regressor.intercept_
slope = linear_regressor.coef_[0]

# Calculate performance metrics
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Print the coefficients and performance metrics
print(f"Intercept (Î²0): {intercept}")
print(f"Slope (Î²1): {slope}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")

# Function to interpret the R-squared value
def interpret_r_squared(r2):
    if r2 < 0 or r2 > 1:
        return "R-squared value must be between 0 and 1."
    interpretation = (f"The R-squared value of {r2:.2f} indicates that approximately {r2 * 100:.2f}% of the variance in the dependent variable "
                      f"can be explained by the independent variable(s).")
    if r2 > 0.9:
        interpretation += " This indicates an excellent fit of the model to the data."
    elif r2 > 0.7:
        interpretation += " This indicates a very good fit of the model to the data."
    elif r2 > 0.5:
        interpretation += " This indicates a good fit of the model to the data."
    elif r2 > 0.3:
        interpretation += " This indicates a moderate fit of the model to the data."
    elif r2 > 0.1:
        interpretation += " This indicates a weak fit of the model to the data."
    else:
        interpretation += " This indicates a very weak fit of the model to the data."
    print(interpretation)

# Interpret the R-squared value
interpret_r_squared(r2)

# Plotting the regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Observed Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression: Shipment Count vs. Mean Transit Time')
plt.xlabel('Shipment Count')
plt.ylabel('Mean Transit Time (days)')
plt.legend()
plt.grid(True)
plt.show()
```
